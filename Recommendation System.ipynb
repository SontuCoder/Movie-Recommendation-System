# Preprocessing of Dataset:-

Separate the columns from total dataset:-

genres, id, keywords, title, overview, 
release_date, status, cast, crew:-

```python
movies = movies[['movie_id','title','overview','genres','keywords','release_date','cast','crew','status']]
```

```python
movies.isnull().sum()
```

Remove missing Data form Dataset:-

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

Calculate duplicate Data form Dataset:-

```python
movies.duplicated().sum
```

Change the formate of genres, keywords, cast, crew columns:-

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
   
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
def fatch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fatch_director)
```

After change the formate of columns, the tables are:-

```python
movies.head()
```

Remove space between two words from keywords, genres, cast, crew:-

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
import re
def combine_columns(row):
    return [row['overview']] + row['genres'] + row['keywords'] + row['cast'] + row['crew']


movies['tags'] = movies.apply(combine_columns, axis=1)
```

```python
movies
```

Convert the "movies" dataset in a new dataset "new_df":-

```python
new_df = movies[['movie_id','title','tags','release_date','status']]
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

# Vectorization:-

```python
import nltk
!pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

# Calculate Cosine distance of all vectors:-

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```
